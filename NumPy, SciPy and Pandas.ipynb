# NumPy, SciPy and Pandas

So let's get started

```python
# some imports and config
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] =  10, 6

import numpy as np
import pandas as pd 
from __future__ import print_function
import statsmodels.api as sm
np.random.seed(9876789)
from scipy import stats 

# http://api.stlouisfed.org/docs/fred/
import fredapi 
fred = fredapi.Fred(api_key='266ecfe55e7ad829f235934dc60c4667')
```

Now let's get some data from st louis fed.

```python
mv = fred.get_series('DGS10').resample('Q', how='mean')
gdp = fred.get_series('FEDFUNDS').resample('Q', how='mean')
```

let's combine these two in a dataframe and plot it

```python
combined = pd.concat([mv, gdp], axis=1) 
combined.columns = ["DGS10", "FEDFUNDS"]

combined.plot()
```

we need to compute first diffrence to introduce stationarity.

```python
dfd = combined.diff().dropna()
dfd.plot()
```

Simple OLS regression

```python
y = dfd["DGS10"].values
x = dfd["FEDFUNDS"].values
result = sm.OLS( x, y ).fit()
result.summary()
```
